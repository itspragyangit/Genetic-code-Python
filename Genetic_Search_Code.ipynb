```
# usual genetic code distribution

genetic_code = { 'phe':['UUU','UUC'], 'leu':['UUA','UUG','CUU','CUC','CUA','CUG'], 'ile' : ['AUU','AUC','AUA'], 'met':['AUG'], 'val':['GUU','GUC','GUA','GUG'], 'ser':['UCU','UCC','UCA','UCG','AGU','AGC'],'pro':['CCU','CCC','CCA','CCG'], 'thr':['ACU','ACC','ACA','ACG'], 'ala':['GCU','GCC','GCA', 'GCG' ], 'tyr':['UAC','UAU'], 'his':['CAU','CAC'],'gln':['CAA','CAG'],'asn':['AAU','AAC'], 'lys':['AAA','AAG'], 'asp':['GAU','GAC'],'glu':['GAA','GAG'], 'cys':['UGU','UGC'], 'trp':['UGG'], 'arg':['CGU','CGC','CGA','CGG'],'gly':['GGU','GGC','GGA','GGG'],'STOP':['UAA','UAG','UGA','AGA','AGG']}
```

```
genetic_code
```

```
# changing a DNA strand to RNA

DNA = "ATCGCGGGTCGATGCTGGACTTGACTGATTCCTGAAAAACGT"     # Enter your DNA Sequence here
RNA = DNA.replace('T','U')
print(RNA,"--> ", len(RNA))
```

```
gcDict = { }
for  protein,rnas in genetic_code.items():
  for rna in rnas:
    gcDict[rna] = protein
print(gcDict)
```

```
# getting the type of proteins

protien = ' '
for i in range(0,len(RNA)-2,3):
  rna = RNA[ i : i+3]
  protien += gcDict[rna] + '-'
protien
```
